```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO  # Importar StringIO para manejar datos en memoria

# 1. Cargar los Datos
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame usando StringIO correctamente
df = pd.read_csv(StringIO(data))

# 2. Conversin de la Columna 'Fecha'
df['fecha'] = pd.to_datetime(df['fecha'])

# 3. Corregir los Grficos
# Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-', color='blue')
plt.title("Ventas Diarias - Grfico de Lneas")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red')
plt.title("Ventas Diarias - Grfico de Dispersin")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# Histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribucin de Ventas - Histograma")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(axis='y')
plt.show()
```
