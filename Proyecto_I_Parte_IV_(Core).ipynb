```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# 1. Cargar los Datos
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame
df = pd.read_csv(StringIO(data))

# 2. Conversi√≥n de la Columna 'Fecha'
df['fecha'] = pd.to_datetime(df['fecha'])

# 3. An√°lisis Descriptivo
print("Estad√≠sticas descriptivas:")
print(df['ventas'].describe())

# 4. Visualizaci√≥n de la Distribuci√≥n: Histograma y Boxplot

# Histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribuci√≥n de Ventas - Histograma")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(axis='y')
plt.show()

# Boxplot de las ventas
plt.figure(figsize=(10, 6))
plt.boxplot(df['ventas'], vert=False, patch_artist=True)
plt.title("Distribuci√≥n de Ventas - Boxplot")
plt.xlabel("Ventas")
plt.show()

# 5. Gr√°fico de L√≠neas para Mostrar Tendencias
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-', color='blue')
plt.title("Ventas Diarias - Gr√°fico de L√≠neas")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# 6. Gr√°fico de Dispersi√≥n para Analizar Relaciones
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red')
plt.title("Ventas Diarias - Gr√°fico de Dispersi√≥n")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()

# 7. Visualizaci√≥n Combinada: Histograma y Boxplot

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Histograma
ax1.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
ax1.set_title("Histograma de Ventas")
ax1.set_xlabel("Ventas")
ax1.set_ylabel("Frecuencia")
ax1.grid(axis='y')

# Boxplot
ax2.boxplot(df['ventas'], vert=False, patch_artist=True)
ax2.set_title("Boxplot de Ventas")
ax2.set_xlabel("Ventas")

plt.show()
```
